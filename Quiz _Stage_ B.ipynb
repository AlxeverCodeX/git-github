```
#13
#we normalise first
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scale_fit = scaler.fit_transform(df)
normalised_data = pd.DataFrame(scale_fit, columns=df.columns)
features_df = normalised_data.drop('Appliances', axis=1)
appliance_target = normalised_data['Appliances']
```

```
#we split into traning and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, appliance_target, test_size=0.3, random_state=42)
```

```

#import LinearRegression() from scikit.learn
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()


linear_model.fit(x_train, y_train)

var_to_get = linear_model.predict(x_test)
```

```

# MAE 
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, var_to_get)
round(mae, 2)
```

```
#14
#RSS
rss = np.sum(np.square(y_test - var_to_get))
round(rss, 2)
```

```
#15
# Root mean square error
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, var_to_get))
round(rmse, 3)
```

```
#16
#Coefficient of determination
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, var_to_get)
round(r2_score, 2)
```

```
#function for the mae, rss, rmse and r2_score
def get_metrics(test, predicted_value):
    
    from sklearn.metrics import r2_score
    from sklearn.metrics import  mean_squared_error as mse
    
    print("Mean absolute error: %.2f" % np.mean(np.absolute(test - predicted_value)))
    print("Residual sum of squares (MSE): %.2f" % np.sum((test - predicted_value) ** 2))
    print("Root mean square error (RMSE): %.3f" % np.sqrt(mse(test, predicted_value)))
    print("R2-score: %.2f" % r2_score(test, predicted_value))
    
```

```
# time to call function we created so we can be clarified
get_metrics(y_test, var_to_get)
```

```
#17

# function for feature weights
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
linear_model_weights.sort_values(by='Linear_Model_Weight')
```
