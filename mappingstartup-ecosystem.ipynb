```python
#merging the two csv
people = pd.read_csv('../input/startup-investments/people.csv', index_col=0)
degrees = pd.read_csv('../input/startup-investments/degrees.csv', index_col=0)

df = people.merge(degrees, on='object_id')
```

```python
# check for missing data in the  data
df.isnull().sum()
```

```python
#lets check percentange of missing value
missing_value_train=pd.DataFrame(data=df.isnull().sum()/len(df)*100,index=df.columns,columns=['missing%'])
missing_value_train.sort_values(by='missing%',ascending=False, inplace=True)
missing_value_train[missing_value_train['missing%'] >0]
```

```python
# Dropping null value
df.drop(['birthplace',"graduated_at"], axis="columns", inplace=True)
```

```python
df.info()
```

```python
#lets check percentange of missing value
missing_value_train=pd.DataFrame(data=df.isnull().sum()/len(df)*100,index=df.columns,columns=['missing%'])
missing_value_train.sort_values(by='missing%',ascending=False, inplace=True)
missing_value_train[missing_value_train['missing%'] >0]
```

```python
#to drop any more missing values
df.dropna(inplace=True)
```

Visualization


```python
#visualization libaries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
def count_plots(df, col_count):
    for i, col in enumerate(df.columns):
        plt.figure(i, figsize=(10,5))
        sns.countplot(x=col, data=df, order=pd.value_counts(df[col]).iloc[:col_count].index)
        plt.xticks(rotation=70)
        
count_columns = df[["full_name","degree_type","subject","institution"]]

count_plots(count_columns, 10)
```
